```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def cosinus(amplitude, frequency, phase, time):
    return amplitude * np.cos(2 * np.pi * frequency * time + phase)

amplitudine = 1

frecventa_1 = 10

faza = 0

time_of_view = 1 

n_samples = int(10e5) 

time_continous = np.linspace(0, time_of_view, n_samples)

signal_continous1 = cosinus(amplitudine, frecventa_1, faza, time_continous)

plt.plot(time_continous, signal_continous1)
plt.title('Fig 1')
plt.show()
```

```python
n = 1000

time_continous = np.linspace(0, time_of_view, n)

signal_continous1 = cosinus(amplitudine, frecventa_1, faza, time_continous)

exponentiala = cosinus(amplitudine, frecventa_1, faza, time_continous) * math.e ** (-2 * np.pi * 1j * time_continous)

plt.plot(exponentiala.real, exponentiala.imag)
plt.title('Fig 1 b')
plt.show()
```

```python
n = 1000

w1 = 1

w2 = 2

w3 = 5

w4 = 7

time_continous = np.linspace(0, time_of_view, n)

signal_continous1 = cosinus(amplitudine, frecventa_1, faza, time_continous)

exponentiala_1 = cosinus(amplitudine, frecventa_1, faza, time_continous) * math.e ** (-2 * np.pi * 1j * time_continous * w1)

exponentiala_2 = cosinus(amplitudine, frecventa_1, faza, time_continous) * math.e ** (-2 * np.pi * 1j * time_continous * w2)

exponentiala_3 = cosinus(amplitudine, frecventa_1, faza, time_continous) * math.e ** (-2 * np.pi * 1j * time_continous * w3)

exponentiala_4 = cosinus(amplitudine, frecventa_1, faza, time_continous) * math.e ** (-2 * np.pi * 1j * time_continous * w4)

plt.plot(exponentiala_1.real, exponentiala_1.imag)
plt.title('w = 1')
plt.show()
plt.plot(exponentiala_2.real, exponentiala_2.imag)
plt.title('w = 2')
plt.show()
plt.plot(exponentiala_3.real, exponentiala_3.imag)
plt.title('w = 5')
plt.show()
plt.plot(exponentiala_4.real, exponentiala_4.imag)
plt.title('w = 7')
plt.show()
```

```python
frecventa_2 = 15

frecventa_3 = 30

frecventa_4 = 45

signal_continous_2 = cosinus(amplitudine, frecventa_2, faza, time_continous) + 2*cosinus(amplitudine, frecventa_3, faza, time_continous) + 0.5*cosinus(amplitudine, frecventa_4, faza, time_continous)

plt.plot(time_continous, signal_continous_2)
plt.title('Fig 3')
plt.show()
```

```python
w_5 = 100

signal_continous_3 = signal_continous_2 * math.e ** (-2 * np.pi * 1j * time_continous * w_5)

plt.plot(time_continous, signal_continous_3)
plt.show()
```

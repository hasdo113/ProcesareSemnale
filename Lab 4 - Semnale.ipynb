```python
import numpy as np
import matplotlib.pyplot as plt

# Ex 1
# a)
# Hz = 1/s
# frec_esantionare = 1 / timp_esantionare

timp_esantionare = 7200 # 2h * 60m * 60s
frec_esantionare = 1/timp_esantionare # 1/7200
```

```python
# b)
timp_temperatura = 24 * 60 * 60 #24h * 60m * 60s pt toata ziua - perioada
frec_tempertaura = 1/timp_temperatura 
# timp_temperatura < timp_esantionare/2 - 24*60*60= 1/86400 < 1/14400
# deci frec_temperatura < frec_esantionare / 2 -> esantionarea este corecta
# intervalul minim?maxim este 
```

```python
# c)
def sinus(amplitude, frequency, phase, time):
    return amplitude * np.sin(2 * np.pi * frequency * time + phase)

amplitudine = 1

faza = np.pi/2

# time of view este orizontul de timp = 1 sapt
time_of_view = 7 * timp_temperatura

n_samples = int(10e5)

time_continous = np.linspace(0, time_of_view, n_samples)

signal_continous = sinus(amplitudine, frec_esantionare, faza, time_continous)

time_distinct = np.linspace(0, time_of_view, int(time_of_view/timp_esantionare))

signal_distinct = sinus(amplitudine, frec_esantionare, faza, time_distinct)

fig, axs = plt.subplots(2)
fig.suptitle('Ex 1 - c)')

axs[0].plot(time_continous, signal_continous)
axs[0].grid("grid")

axs[1].stem(time_distinct, signal_distinct)
axs[1].grid("grid")
```

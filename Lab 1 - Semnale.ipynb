```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage import io

# 1 : a) si b)

def cosinus(amplitude, frequency, phase, time):
    return amplitude * np.cos(2 * np.pi * frequency * time + phase)

amplitudine = 1

frecventa1 = 260
faza1 = np.pi/3

frecventa2 = 140
faza2 = -np.pi/3

frecventa3 = 60
faza3 = np.pi/3

time_of_view = 0.03 # orizontul de timp
n_samples = int(10e5) # dim vectorului, numarul de esantioane
time_continous = np.linspace(0, time_of_view, n_samples)# genereaza un vector dupa parametri

signal_continous1 = cosinus(amplitudine, frecventa1, faza1, time_continous)
signal_continous2 = cosinus(amplitudine, frecventa2, faza2, time_continous)
signal_continous3 = cosinus(amplitudine, frecventa3, faza3, time_continous)

fig, axs = plt.subplots(10)
fig.suptitle('Ex 1 & 2')

axs[0].plot(time_continous, signal_continous1)
axs[0].grid("grid")

axs[1].plot(time_continous, signal_continous2)
axs[1].grid("grid")

axs[2].plot(time_continous, signal_continous3)
axs[2].grid("grid")

# c)
# Ts = 1/200, n_samples = 0.03/1/200 = 0.03/0.005

n_samples2 = int(0.03/0.005)

time_distinct = np.linspace(0, time_of_view, n_samples2)

signal_distinct1 = cosinus(amplitudine, frecventa1, faza1, time_distinct)
signal_distinct2 = cosinus(amplitudine, frecventa2, faza2, time_distinct)
signal_distinct3 = cosinus(amplitudine, frecventa3, faza3, time_distinct)

axs[3].stem(time_distinct, signal_distinct1)
axs[3].grid("grid")

axs[4].stem(time_distinct, signal_distinct2)
axs[4].grid("grid")

axs[5].stem(time_distinct, signal_distinct3)
axs[5].grid("grid")

# 2 : a)

frecventa_2a = 400

n_samples_2a = int(1600)
time_2a = np.linspace(0, time_of_view, n_samples_2a)

signal_continous_2a = cosinus(amplitudine, frecventa_2a, 0, time_2a)

axs[6].plot(time_2a, signal_continous_2a)
axs[6].grid("grid")

# 2 : b)

frecventa_2b = 800

time_of_view_2b = 3

time_2b = np.linspace(0, time_of_view_2b, n_samples_2a) #time, 1/frec

signal_continous_2b = cosinus(amplitudine, frecventa_2b, 0, time_2b)

axs[7].plot(time_2b, signal_continous_2b)
axs[7].grid("grid")

# 2 : c) 

def sawtoothSignal(t):
    return np.mod(t, 1./240)

timp = np.linspace(0, 0.05, int(60+1))

axs[8].plot(timp, sawtoothSignal(timp)) 
axs[8].grid("grid")

# 2 : d)

frecventa_2d = 300

signal_continous_2d = cosinus(amplitudine, frecventa_2d, faza1, time_continous)

square = np.floor(signal_continous_2d)

axs[9].plot(time_continous, square)
axs[9].grid("grid")
```

```python
# Ex 1
import numpy as np
import matplotlib.pyplot as plt 

def cosinus(amplitude, frequency, phase, time):
    return amplitude * np.cos(2 * np.pi * frequency * time + phase)

def sinus(amplitude, frequency, phase, time):
    return amplitude * np.sin(2 * np.pi * frequency * time + phase)

amplitudine = 4

frecventa1 = 60

faza1 = 0

time_of_view = 0.03 # orizontul de timp

n_samples = int(10e5) # dim vectorului, numarul de esantioane

time_continous = np.linspace(0, time_of_view, n_samples)# genereaza un vector dupa parametri

signal_continous1 = sinus(amplitudine, frecventa1, faza1, time_continous)

signal_continous2 = cosinus(amplitudine, frecventa1, faza1, time_continous)

fig, axs = plt.subplots(2)
fig.suptitle('Ex 1')

axs[0].plot(time_continous, signal_continous1)
axs[0].grid("grid")

axs[1].plot(time_continous, signal_continous2)
axs[1].grid("grid")
```

```python
#Ex 1 & 2
import numpy as np
import matplotlib.pyplot as plt 

def cosinus(amplitude, frequency, phase, time):
    return amplitude * np.cos(2 * np.pi * frequency * time + phase)

amplitudine = 1

frecventa_1 = 400

frecventa_2 = 150

frecventa_3 = 500

frecventa_4 = 600

faza = 0

time_of_view = 0.03 

n_samples = int(10e5) 

time_continous = np.linspace(0, time_of_view, n_samples)

signal_continous1 = cosinus(amplitudine, frecventa_1, faza, time_continous)

signal_continous2 = cosinus(amplitudine, frecventa_2, faza, time_continous)

signal_continous3 = cosinus(amplitudine, frecventa_3, faza, time_continous)

signal_continous4 = cosinus(amplitudine, frecventa_4, faza, time_continous)

n_samples2 = int(6)

time_distinct = np.linspace(0, time_of_view, n_samples2 + 1)

signal_distinct1 = cosinus(amplitudine, 400, faza, time_distinct)

signal_distinct2 = cosinus(amplitudine, 150, faza, time_distinct)

fig, axs = plt.subplots(4)
fig.suptitle('Ex 1 & 2')

axs[0].plot(time_continous, signal_continous1)
axs[0].stem(time_distinct, signal_distinct1)
axs[0].grid("grid")

axs[1].plot(time_continous, signal_continous2)
axs[1].stem(time_distinct, signal_distinct2)
axs[1].grid("grid")

axs[2].plot(time_continous, signal_continous3)
axs[2].grid("grid")

axs[3].plot(time_continous, signal_continous4)
axs[3].grid("grid")
```
